```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.poly1d([-1/48**2, 0, 1])
```

```python
x = np.linspace(0, 48, 100)
plt.figure(figsize=(6,3))
plt.plot(x, y(x))
plt.plot([12, 12], [0, y(12)], color='black')
plt.plot([24, 24], [0, y(24)], color='black')
plt.plot([36, 36], [0, y(36)], color='black')
plt.plot([12, 24, 36], [y(12), y(24), y(36)], 'o')
plt.annotate('  {0:.2f}'.format(y(12)), xy=(12, y(12)))
plt.annotate('  {0:.2f}'.format(y(24)), xy=(24, y(24)))
plt.annotate('  {0:.2f}'.format(y(36)), xy=(36, y(36)))
plt.ylim(0, plt.ylim()[1])
plt.xticks([0, 6, 12, 18, 24, 30, 36, 42, 48])
plt.xlim(0, plt.xlim()[1])
plt.ylabel('Percent of total points')
plt.xlabel('Hours late')
plt.tight_layout()
plt.savefig('img/late_policy.png', bbox_inches='tight')
```
